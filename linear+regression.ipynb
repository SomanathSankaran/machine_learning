```python
import numpy as np
```

```python
x=np.array([2,4,6,8,10])
y=np.array(x*3)
```

```python
#cost function
```

```python
def cost_function(X,Y,m,c,alpha=0.1):
    
    assert(len(X)==len(Y) and (len(X)==len(m)) and (len(c)==len(m)))
    h=np.dot(m,X)+c
    error_square=np.square(Y-h)
    print("error",sum(error_square),"h:",h)
    if sum(error_square)==0:
        pass
    else:
        c=c-alpha*(np.dot(m,X)+c-Y)
        m=m-alpha*np.dot((np.dot(m,X)+c-Y),X)
        cost_function(X,Y,m,c,alpha=0.1)
    
    
```

```python
cost_function(x,y,np.ones_like(x),np.zeros_like(x),1)
```

```python
cost_function(x,y,np.ones_like(x),np.zeros_like(x),0.5)
```

```python
cost_function(x,y,np.ones_like(x),np.zeros_like(x),1)
```
